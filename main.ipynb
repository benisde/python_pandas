```python
# Averaging sharpe ratios
whale_sharpe_avg = whale_sharpe_ratios.mean()
algo_sharpe_avg = algo_sharpe_ratios.mean()

# Print average Sharpe Ratios of each portfolio
print (f"Whale's sharpe ratio average: {whale_sharpe_avg}")
print (f"Algorithmic's sharpe ratio average: {algo_sharpe_avg}")
```

```python
# Compare if Algorithmic strategies outperform S&P TSX
algo_sharpe_avg > sp_tsx_sharp_avg
```
